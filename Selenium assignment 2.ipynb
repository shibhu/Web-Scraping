```python

i=0
time.sleep(2)
url=[]
# getting all the product URL's
while i<2:
    
    ul = driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
    for r in ul:
        url.append(r.get_attribute("href"))
    i+=1
    try: 
        nt = driver.find_element_by_xpath("//li[@class='a-last']")
        nt.click()
    except:
        nt = driver.find_element_by_xpath("//*[@id='search']/div[1]/div/div[1]/div/span[3]/div[2]/div[54]/span/div/div/ul/li[7]/a")
        next_url = nt.get_attribute("href")
        driver.get(next_url)
    time.sleep(2)
```

```python
# creating lists of the required data
n=[]
price=[]
rating=[]
total_rating=[]
delivery=[]
availability=[]
brand=[]
returndel=[]
other=[]
# fetching the data from the product URL's
for i in url[:]:
    driver.get(i)
    time.sleep(2)

    try:
        n.append(driver.find_element_by_xpath("//h1[@class='a-size-large a-spacing-none']").text)
    except NoSuchElementException:
        n.append("-")
    try:
        price.append(driver.find_element_by_xpath("//span[@id='priceblock_ourprice']").text)
    except NoSuchElementException:
        price.append("-")
    try:
        rating.append(driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']").text)
    except NoSuchElementException:
        rating.append("-")
    try:
        total_rating.append(driver.find_element_by_xpath("//*[@id='reviewsMedley']/div/div[1]/div[2]/div[2]/span").text)
    except NoSuchElementException:
        total_rating.append("-")
    try:
        delivery.append(driver.find_element_by_xpath("//*[@id='ddmDeliveryMessage']/b").text)
    except NoSuchElementException:
        delivery.append("-")    
    try:
        availability.append(driver.find_element_by_xpath("//*[@id='availability']/span").text)
    except NoSuchElementException:
        availability.append("-") 
    try:
        brand.append(driver.find_element_by_xpath("//*[@id='productOverview_feature_div']/div/table/tbody/tr[1]/td[2]/span").text)
    except NoSuchElementException:
        brand.append("-") 
    try:
        returndel.append(driver.find_element_by_xpath("//*[@id='RETURNS_POLICY']/span/div[2]/a").text)
    except NoSuchElementException:
        returndel.append("-")
    try:
        other.append(driver.find_element_by_xpath("//*[@id='feature-bullets']").text)
    except NoSuchElementException:
        other.append("-")
```

```python
# creating the dataframe
amazon = pd.DataFrame({
            "name":n,
"price":price,
"rating":rating,
"total_ratings":total_rating,
"delivery":delivery,
"availability":availability,
"brand":brand,
"return":returndel,
"other":other,
"URL":url
        })
amazon.head()
```

```python
#saving to csv file
amazon.to_csv("Amazon.csv")
```

```python
# fetching the data
quarter=[50,49]
for quat in quarter:
    flag=0
    while flag==0:
        for i in range(1,11):
            try:
                date.append(driver.find_element_by_xpath("//*[@id='tablepress-"+str(quat)+"']/tbody/tr["+ str(i)+"]/td[2]").text)
                name.append(driver.find_element_by_xpath("//*[@id='tablepress-"+str(quat)+"']/tbody/tr["+ str(i)+"]/td[3]").text)
                vertical.append(driver.find_element_by_xpath("//*[@id='tablepress-"+str(quat)+"']/tbody/tr["+ str(i)+"]/td[4]").text)
                subvertical.append(driver.find_element_by_xpath("//*[@id='tablepress-"+str(quat)+"']/tbody/tr["+ str(i)+"]/td[5]").text)
                location.append(driver.find_element_by_xpath("//*[@id='tablepress-"+str(quat)+"']/tbody/tr["+ str(i)+"]/td[6]").text)
                investor.append(driver.find_element_by_xpath("//*[@id='tablepress-"+str(quat)+"']/tbody/tr["+ str(i)+"]/td[7]").text)
                investtype.append(driver.find_element_by_xpath("//*[@id='tablepress-"+str(quat)+"']/tbody/tr["+ str(i)+"]/td[8]").text)
                amount.append(driver.find_element_by_xpath("//*[@id='tablepress-"+str(quat)+"']/tbody/tr["+ str(i)+"]/td[9]").text)
            except:
                flag=1
                break
        try:
            button = driver.find_element_by_xpath("//*[@id='tablepress-"+str(quat)+"_next']")
            button.click()
            time.sleep(2)
        except:
            break
            
```

```python
# fetching the URL
url = "https://www.hostelworld.com/"
driver.get(url)
```

```python
# searching the city
import time
search = driver.find_element_by_xpath("//*[@id='__layout']/div/div[1]/div[1]/div/div[2]/div[4]/div/div[2]/div/div[1]/div")
search.click()
search = driver.find_element_by_xpath("//*[@id='search-input-field']")
search.send_keys("London")

time.sleep(5)
search_bt = driver.find_element_by_xpath("//*[@id='predicted-search-results']/li[2]")
search_bt.click()
time.sleep(3)
search_bt = driver.find_element_by_xpath("//*[@id='search-button']")
search_bt.click()
time.sleep(4)
```

```python
# creating the list of elements
import time
from selenium.common.exceptions import NoSuchElementException
name=[]
distance=[]
rating=[]
treviews=[]
oreviews=[]
pprice=[]
dprice=[]
discription=[]
facilities=[]
url=[]

# fetching the elements
while True:
    for i in driver.find_elements_by_xpath("//*[@class='title title-6']/a"):
        try:
            name.append(i.text)
        except NoSuchElementException:
            name.append("-")
    for i in driver.find_elements_by_xpath("//span[@class='description']"):
        try:
            distance.append(i.text)
        except NoSuchElementException:
            distance.append("-")
    for i in range(3,33):
        try:
            pprice.append(driver.find_element_by_xpath("//*[@id='__layout']/div/div[1]/div[4]/div/div/div[" + str(i) + "]/div[3]/a[1]/div[1]/div").text)
        except NoSuchElementException:
            pprice.append("-")
        try:
            dprice.append(driver.find_element_by_xpath("//*[@id='__layout']/div/div[1]/div[4]/div/div/div[" + str(i) + "]/div[3]/a[1]/div[2]/div").text)
        except NoSuchElementException:
            dprice.append("-")
    for i in driver.find_elements_by_xpath("//*[@class='facilities-label facilities']"):
        facilities.append(i.text)
    
    ul = driver.find_elements_by_xpath("//*[@class='title title-6']/a")
    for i in ul:
        url.append(i.get_attribute("href"))
    
    try:
        search = driver.find_element_by_xpath("//div[@class='pagination-item pagination-next']")
        search.click()
        time.sleep(5)
    except:
        break;
        
# fetching the elements by opening the places
for i in url:
        driver.get(i)
        time.sleep(2)
        try:
            rating.append(driver.find_element_by_xpath("//*[@id='__layout']/div/div[1]/section/div[6]/div/div[1]/div[1]/div[1]").text)
        except NoSuchElementException:
            rating.append("-")
        try:
            treviews.append(driver.find_element_by_xpath("//*[@id='__layout']/div/div[1]/section/div[6]/div/div[1]/div[1]/div[1]").text)
        except NoSuchElementException:
            treviews.append("-")
        try:
            oreviews.append(driver.find_element_by_xpath("//*[@id='__layout']/div/div[1]/section/div[6]/div/div[1]/div[1]/div[2]/div[1]/span").text)
        except NoSuchElementException:
            oreviews.append("-")    
        try: 
            exp=driver.find_element_by_xpath("//*[@id='__layout']/div/div[1]/section/div[6]/div/div[2]/div/div/a")
            exp.click()
        except:
            try:
                discription.append(driver.find_element_by_xpath("//*[@id='__layout']/div/div[1]/section/div[6]/div/div[2]/div/div/div[2]").text)
            except NoSuchElementException:
                discription.append("-")  
        try:
            discription.append(driver.find_element_by_xpath("//*[@id='__layout']/div/div[1]/section/div[6]/div/div[2]/div/div/div[2]").text)
        except NoSuchElementException:
            discription.append("-") 
```

```python
discription=list(set(discription))
```

```python
# creating the dataframe
data = pd.DataFrame({
"name":name,
"distance":distance,
"rating":rating,
"treviews":treviews,
"oreviews":oreviews,
"pprice":pprice,
"dprice":dprice,
"discription":discription,
"facilities":facilities
})
data.head()
```
