```python
# getting all the elements XPATH
team = driver.find_elements_by_xpath("//div[@class='fixture__teams']")
fixture = driver.find_elements_by_xpath("//strong[@class='fixture__name fixture__name--with-margin']")
address = driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span")
date = driver.find_elements_by_xpath("//span[@class='fixture__date']")
month = driver.find_elements_by_xpath("//span[@class='fixture__month']")
time = driver.find_elements_by_xpath("//span[@class='fixture__time']")
```

```python
# creating the dataframe
teams = pd.DataFrame({
    "team":teams,
    "fixture":fixtures,
    "address":addresses,
    "date":dates,
    "month":months,
    "time":times
})
teams.head()
```

```python
links = driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
# creating list to store the data
url=[]
projects=[]
about=[]
contributor=[]
language=[]
for i in links:
    projects.append(i.text)
for i in links:
    url.append(i.get_attribute('href'))
```

```python
#looping through all the elements for the required tags
for ul in url:
    driver.get(ul)
    time.sleep(2)
    try:
        about.append(driver.find_element_by_xpath("//p[@class='f4 mt-3']").text)
    except NoSuchElementException:
        about.append("-")
    try:
        contributor.append(driver.find_element_by_xpath("//*[text()[contains(., 'Contributors')]]").text)
    except NoSuchElementException:
        contributor.append("-")
    try:
        temp=driver.find_element_by_xpath("//*[text()[contains(., 'Languages')]]").text
        x = driver.find_elements_by_xpath("//ul[@class='list-style-none']")
        language.append(x[1].text)
    except NoSuchElementException:
        language.append("-")
    
    
```

```python
# creating the dataframe
github = pd.DataFrame({
'project name':projects,
'summary':about,
'contributor':contributor,
'languages':language
})
github.head()
```

```python
# opening the portal
url = "https://www.imdb.com/list/ls095964455/"
driver.get(url)
```

```python
# creating the list to collect the data
name=[]
year=[]
genre=[]
runtime=[]
ratings=[]
votes=[]
# looping through each element to collect data
for i in driver.find_elements_by_xpath("//h3[@class='lister-item-header']/a"):
    name.append(i.text)
for i in driver.find_elements_by_xpath("//h3[@class='lister-item-header']/span[2]"):
    year.append(i.text)
for i in driver.find_elements_by_xpath("//p[@class='text-muted text-small']/span[3]"):
    runtime.append(i.text)
for i in driver.find_elements_by_xpath("//p[@class='text-muted text-small']/span[5]"):
    genre.append(i.text)
for i in driver.find_elements_by_xpath("//div[@class='lister-item-content']/p[4]/span[2]"):
    votes.append(i.text)
for i in driver.find_elements_by_xpath("//div[@class='ipl-rating-widget']/div/span[2]"):
    ratings.append(i.text)
# creating the dataset
movies=pd.DataFrame({
    "name":name,
    "year":year,
    "genre":genre,
    "runtime":runtime,
    "ratings":ratings,
    "votes":votes
})
movies.head()
```

```python
# clicking the text to reach the destionation portal
button = driver.find_element_by_xpath("//b[text()[contains(., 'View ALL Data Sets')]]")
button.click()
```

```python
# creating the list ot save data
dataname=[]
datatypes=[]
task=[]
attrtype=[]
instances=[]
attribute=[]
year=[]
# looping through each element to collect the data
for i in range(2,560):
    dataname.append(driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str(i)+"]/td[1]").text)
    datatypes.append(driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str(i)+"]/td[2]").text)
    task.append(driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str(i)+"]/td[3]").text)
    attrtype.append(driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str(i)+"]/td[4]").text)
    instances.append(driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str(i)+"]/td[5]").text)
    attribute.append(driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str(i)+"]/td[6]").text)
    year.append(driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str(i)+"]/td[7]").text)
# creating the dataset
dataset=pd.DataFrame({
    "Name":dataname,
    "Data types":datatypes,
    "Task":task,
    "attributes type":attrtype,
    "Instances":instances,
    "Attribute":attribute,
    "Year":year
})
dataset.head()
```
